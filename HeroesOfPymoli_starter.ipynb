```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_palyer = purchase_data["SN"].nunique()
Players =pd.DataFrame({"Total Player": total_palyer}, index=[0])
Players.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = purchase_data["Item ID"].nunique() # nunique gets us the count of unique Item ID
average_price = purchase_data["Price"].mean()
number_of_purchase = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

#creating a DataFrame

summary_data = pd.DataFrame({"Number of Unique Items ": unique_item,
                             "Average Price": average_price,
                             "Number of Purchases" : number_of_purchase,
                             "Total Revenue" : total_revenue}, index=[0])
summary_data.round(2) # .round(2) is used for rounding up the decimal by 2 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
df =purchase_data[["Purchase ID","SN","Age","Gender","Price"]]
```

```python
new_df =df.drop_duplicates(["SN"], keep = "first")
```

```python
total_count =new_df["Gender"].value_counts()
```

```python
percent_player = new_df["Gender"].value_counts(normalize=True) * 100
```

```python
gender_demograpic_summary = pd.DataFrame({"Total Count": total_count,
                               "Percentage of Players": percent_player })
gender_demograpic_summary.round(2)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = df["Gender"].value_counts()
```

```python
avg_purchase_price = df.groupby(["Gender"])["Price"].mean()
```

```python
total_purchase_value = df.groupby(["Gender"])["Price"].sum()
```

```python
avg_per_person = df.groupby(["Gender"])["Price"].sum()/new_df["Gender"].value_counts()
```

```python
Purchase_analysis = pd.DataFrame({"Purchase Count" : purchase_count,
                                     "Average Purchase Price" : avg_purchase_price,
                                     "Total Purchase Value" : total_purchase_value,
                                     "Avg Total Purchase per Person": avg_per_person})
Purchase_analysis.round(2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9, 14, 19, 24, 29, 34, 39, 100]
names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
new_df["Summary by Age"] = pd.cut(new_df["Age"], bins, labels=names)
```

```python
age_count = new_df.groupby(["Summary by Age"])["Age"].count()
```

```python
age_percent = age_count/new_df["Age"].count()*100
```

## Purchasing Analysis (Age)

```python
age_demo_summary = pd.DataFrame({"Total Count": age_count,
                                 "Percentage of players": age_percent})
age_demo_summary.round(2)
```

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
bins = [0,9, 14, 19, 24, 29, 34, 39, 100]
names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
df["Age Summary"] = pd.cut(df["Age"], bins, labels=names)
```

```python
avg_purchase_price = df.groupby(["Age Summary"])["Price"].mean()
```

```python
total_purchase_value = df.groupby(["Age Summary"])["Price"].sum()
```

```python
avg_per_person = df.groupby(["Age Summary"])["Price"].sum()/new_df.groupby(["Summary by Age"])["Age"].count()
```

```python
Age_Purchase_analysis = pd.DataFrame({"Purchase Count" : age_count,
                                     "Average Purchase Price" : avg_purchase_price,
                                     "Total Purchase Value" : total_purchase_value,
                                     "Avg Total Purchase per Person": avg_per_person})
Age_Purchase_analysis.round(2)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item_df = purchase_data[["Item ID","Item Name", "Price"]]
```

```python
purchase_count = popular_item_df.groupby(["Item ID","Item Name"])["Item ID"].count()
```

```python
item_price = popular_item_df.groupby(["Item ID","Item Name"])["Price"].max() # .max will get max value from every cell in groupby 
```

```python
total_purchase_value = popular_item_df.groupby(["Item ID","Item Name"])["Price"].sum()
```

```python
popular_item_summary = pd.DataFrame({"Total Count":purchase_count,
                              "Item Price":item_price,
                              "Total Purchase Value": total_purchase_value})
```

```python
popular_item_summary.sort_values(by= ["Total Count"],ascending=False).round(2)
```
